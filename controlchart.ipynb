```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('updated_csv_file.csv');
data
```

```python
mean = data['first_sexual_intercourse'].mean()
std_dev = data['first_sexual_intercourse'].std()
```

```python
ucl = mean + 3 * std_dev
lcl = mean - 3 * std_dev
```

```python
fig, ax = plt.subplots()
ax.plot(data['first_sexual_intercourse'], 'o', label='Data')
ax.axhline(y=mean, color='r', label='Mean')
ax.axhline(y=ucl, color='g', linestyle='--', label='UCL')
ax.axhline(y=lcl, color='g', linestyle='--', label='LCL')
ax.set_xlabel('Sample Number')
ax.set_ylabel('Value')
ax.legend()
plt.show()
```

```python
fig, bx = plt.subplots()
bx.plot(data['no_of_pregnancies'], 'o', label='Data')
bx.axhline(y=mean, color='r', label='Mean')
bx.axhline(y=ucl, color='g', linestyle='--', label='UCL')
bx.axhline(y=lcl, color='g', linestyle='--', label='LCL')
bx.set_xlabel('No of Pregencies')
bx.set_ylabel('Value')
bx.legend()
plt.show()
```
